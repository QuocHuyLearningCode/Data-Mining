```python
# Authors: The scikit-learn developers
# SPDX-License-Identifier: BSD-3-Clause
import matplotlib.pyplot as plt  # Thư viện vẽ đồ thị
import numpy as np  # Thư viện tính toán số học
from matplotlib.colors import ListedColormap  # Dùng để tạo bảng màu

# Import các module trong scikit-learn
from sklearn.datasets import make_circles, make_classification, make_moons  # Tạo các tập dữ liệu
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis  # Phân tích phân biệt QDA
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier  # Thuật toán tập hợp
from sklearn.gaussian_process import GaussianProcessClassifier  # Bộ phân loại Gaussian Process
from sklearn.gaussian_process.kernels import RBF  # Kernel RBF
from sklearn.inspection import DecisionBoundaryDisplay  # Hiển thị ranh giới quyết định
from sklearn.model_selection import train_test_split  # Chia tập dữ liệu thành tập huấn luyện và kiểm tra
from sklearn.naive_bayes import GaussianNB  # Phân loại Naive Bayes
from sklearn.neighbors import KNeighborsClassifier  # Phân loại K-Nearest Neighbors
from sklearn.neural_network import MLPClassifier  # Mạng nơ-ron
from sklearn.pipeline import make_pipeline  # Xây dựng pipeline
from sklearn.preprocessing import StandardScaler  # Chuẩn hóa dữ liệu
from sklearn.svm import SVC  # Máy vector hỗ trợ
from sklearn.tree import DecisionTreeClassifier  # Cây quyết định

# Danh sách tên các mô hình
names = [
    "Nearest Neighbors",  # Láng giềng gần nhất
    "Linear SVM",  # SVM tuyến tính
    "RBF SVM",  # SVM với kernel RBF
    "Gaussian Process",  # Quá trình Gaussian
    "Decision Tree",  # Cây quyết định
    "Random Forest",  # Rừng ngẫu nhiên
    "Neural Net",  # Mạng nơ-ron
    "AdaBoost",  # Thuật toán AdaBoost
    "Naive Bayes",  # Naive Bayes
    "QDA",  # Phân tích phân biệt QDA
]

# Tạo danh sách các bộ phân loại
classifiers = [
    KNeighborsClassifier(3),  # Láng giềng gần nhất với K=3
    SVC(kernel="linear", C=0.025, random_state=42),  # SVM tuyến tính
    SVC(gamma=2, C=1, random_state=42),  # SVM với kernel RBF
    GaussianProcessClassifier(1.0 * RBF(1.0), random_state=42),  # Quá trình Gaussian
    DecisionTreeClassifier(max_depth=5, random_state=42),  # Cây quyết định với độ sâu tối đa 5
    RandomForestClassifier(
        max_depth=5, n_estimators=10, max_features=1, random_state=42
    ),  # Rừng ngẫu nhiên
    MLPClassifier(alpha=1, max_iter=1000, random_state=42),  # Mạng nơ-ron
    AdaBoostClassifier(random_state=42),  # Thuật toán AdaBoost
    GaussianNB(),  # Naive Bayes
    QuadraticDiscriminantAnalysis(),  # QDA
]

# Tạo tập dữ liệu phân loại tuyến tính
X, y = make_classification(
    n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1
)
rng = np.random.RandomState(2)  # Tạo ngẫu nhiên
X += 2 * rng.uniform(size=X.shape)  # Thêm nhiễu vào dữ liệu
linearly_separable = (X, y)

# Các tập dữ liệu khác nhau
datasets = [
    make_moons(noise=0.3, random_state=0),  # Tập dữ liệu hình mặt trăng
    make_circles(noise=0.2, factor=0.5, random_state=1),  # Tập dữ liệu hình vòng tròn
    linearly_separable,  # Tập dữ liệu tuyến tính
]

# Vẽ biểu đồ
figure = plt.figure(figsize=(27, 9))  # Đặt kích thước hình
i = 1  # Biến đếm cho các đồ thị

# iterate over datasets
# Lặp qua các tập dữ liệu
for ds_cnt, ds in enumerate(datasets):
    X, y = ds  # Lấy dữ liệu và nhãn
    # Chia dữ liệu thành tập huấn luyện và kiểm tra
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.4, random_state=42
    )

    # Xác định giới hạn đồ thị
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

    # Vẽ tập dữ liệu
    cm = plt.cm.RdBu  # Bảng màu
    cm_bright = ListedColormap(["#FF0000", "#0000FF"])  # Màu sáng
    ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
    if ds_cnt == 0:
        ax.set_title("Input data")  # Tiêu đề cột
    # Vẽ các điểm huấn luyện
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k")
    # Vẽ các điểm kiểm tra
    ax.scatter(
        X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6, edgecolors="k"
    )
    ax.set_xlim(x_min, x_max)  # Giới hạn trục x
    ax.set_ylim(y_min, y_max)  # Giới hạn trục y
    ax.set_xticks(())  # Tắt nhãn trục x
    ax.set_yticks(())  # Tắt nhãn trục y
    i += 1

    # Lặp qua các bộ phân loại
    for name, clf in zip(names, classifiers):
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)

        # Xây dựng pipeline
        clf = make_pipeline(StandardScaler(), clf)
        clf.fit(X_train, y_train)  # Huấn luyện mô hình
        score = clf.score(X_test, y_test)  # Đánh giá mô hình
        DecisionBoundaryDisplay.from_estimator(
            clf, X, cmap=cm, alpha=0.8, ax=ax, eps=0.5
        )  # Vẽ ranh giới quyết định

        # Vẽ lại các điểm dữ liệu
        ax.scatter(
            X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k"
        )
        ax.scatter(
            X_test[:, 0],
            X_test[:, 1],
            c=y_test,
            cmap=cm_bright,
            edgecolors="k",
            alpha=0.6,
        )

        ax.set_xlim(x_min, x_max)  # Giới hạn trục x
        ax.set_ylim(y_min, y_max)  # Giới hạn trục y
        ax.set_xticks(())  # Tắt nhãn trục x
        ax.set_yticks(())  # Tắt nhãn trục y
        if ds_cnt == 0:
            ax.set_title(name)  # Tiêu đề tên mô hình
        ax.text(
            x_max - 0.3,
            y_min + 0.3,
            ("%.2f" % score).lstrip("0"),  # Hiển thị điểm số
            size=15,
            horizontalalignment="right",
        )
        i += 1

plt.tight_layout()  # Tự động điều chỉnh bố cục
plt.show()  # Hiển thị đồ thị
```
