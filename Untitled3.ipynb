```python
import datetime
# chuyển đổi ddmmyyyy sang dd/mm/yyyy
def convert_date(date_str):
    day = date_str[:2]
    month = date_str[2:4]
    year = date_str[4:]
    return f"{day}/{month}/{year}"
    
# chuyển đổi yyyy/mm/yy sang dd/mm/yyyy 
def reversed_date(date_str):
    # Tách chuỗi ngày tháng năm thành danh sách
    parts = date_str.split('/')
    # Đảo ngược thứ tự và kết nối lại với dấu "/"
    reversed_date = '/'.join(parts[::-1])
    return reversed_date

def split_date(date_str):
    # Tách chuỗi thành ngày, tháng, năm
    parts = date_str.split('/')
    
    # Kiểm tra xem danh sách parts có ít nhất ba phần tử không
    if len(parts) >= 3:
        year = parts[0]
        month = parts[1]
        day = parts[2]

        # Kiểm tra năm có đúng định dạng không
        return len(year)
    else:
        return None  # Trả về None nếu không đủ phần tử trong danh sách parts

# chuyển đổi dd-mm-yyyy sang dd/mm/yyyy
def parts_date(date_str):
    # Tách chuỗi ngày tháng năm thành danh sách
    parts = date_str.split('-')
    # Đảo ngược thứ tự và kết nối lại với dấu "/"
    reversed_date = '/'.join(parts[::-1])
    return reversed_date   

# chuyển đôi mm/dd/yyyy sang dd/mm/yyyy
def month_date(date_str):
    # Tách chuỗi ngày tháng năm thành danh sách
    parts = date_str.split('/')
    # Đảo ngược thứ tự và kết nối lại với dấu "/"
    reversed_date = parts[1] + '/' + parts[0] + '/' + parts[2]
    return reversed_date    
def is_valid_date(date_str):
    try:
        parts = date_str.split('/')
        if len(parts) != 3:
            return False
        year, month, day = map(int, parts)
        if month < 1 or month > 12 or day < 1 or day > 31 or year < 1:
            return False
        if month in [4, 6, 9, 11] and day > 30:
            return False
        if month == 2:
            if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
                if day > 29:
                    return False
            elif day > 28:
                return False
        return True
    except ValueError:
        return False
data['Birthday'] = data['Birthday'].str.replace('-', '')  # Xóa tất cả các ký tự không phải số
for index, row in data.iterrows():    

    if len(row['Birthday']) == 8:
        data.at[index, 'Birthday'] = convert_date(row['Birthday'])
    if split_date(row['Birthday']) == 4:
        data.at[index, 'Birthday'] = reversed_date(row['Birthday'])
    if len(row['Birthday']) < 8 or is_valid_date(row['Birthday']) == False :
        current_date = datetime.datetime.now().strftime("%d/%m/%Y")
        data.at[index, 'Birthday'] = current_date
```

```python
# Hàm tính trung bình với xử lý giá trị không phải số
def calculate_mean(column):
    # Lọc ra các giá trị là số
    numeric_values = [float(value) for value in column if str(value).replace('.', '', 1).isdigit()]
    # Tính trung bình của các giá trị số
    if numeric_values:
        return sum(numeric_values) / len(numeric_values)
    else:
        return None

# Hàm tính độ lệch chuẩn với xử lý giá trị không phải số
def calculate_standard_deviation(column):
    mean = calculate_mean(column)
    # Lọc ra các giá trị là số
    numeric_values = [float(value) for value in column if str(value).replace('.', '', 1).isdigit()]
    # Tính độ lệch chuẩn của các giá trị số
    if numeric_values:
        squared_diff = [(x - mean) ** 2 for x in numeric_values]
        variance = sum(squared_diff) / len(numeric_values)
        return variance ** 0.5
    else:
        return None


# Hàm tính giá trị lớn nhất
def calculate_max(column):
    return max(column)

# Hàm tính giá trị nhỏ nhất
def calculate_min(column):
    return min(column)

# Hàm tính phân vị với xử lý giá trị không phải số
def calculate_quantile(column, q):
    # Lọc ra các giá trị là số
    numeric_values = [float(value) for value in column if str(value).replace('.', '', 1).isdigit()]
    # Tính phân vị của các giá trị số
    if numeric_values:
        sorted_column = sorted(numeric_values)
        index = q * (len(sorted_column) - 1)
        if index.is_integer():
            return sorted_column[int(index)]
        else:
            lower_index = int(index // 1)
            upper_index = lower_index + 1
            lower_value = float(sorted_column[lower_index]) # Chuyển đổi thành số
            upper_value = float(sorted_column[upper_index]) # Chuyển đổi thành số
            return (lower_value + upper_value) / 2
    else:
        return None


# Hàm hiển thị thông tin thống kê cơ bản
def display_basic_stats(data):
    for column_name in data.columns:
        column = data[column_name]
        print(f"Column: {column_name}")
        print(f"Mean: {calculate_mean(column)}")
        print(f"Standard Deviation: {calculate_standard_deviation(column)}")
        print(f"Max: {calculate_max(column)}")
        print(f"Min: {calculate_min(column)}")
        print(f"25th Quantile: {calculate_quantile(column, 0.25)}")
        print(f"50th Quantile: {calculate_quantile(column, 0.5)}")
        print(f"75th Quantile: {calculate_quantile(column, 0.75)}")
        print("\n")

# Hiển thị thông tin thống kê cơ bản
display_basic_stats(data)
```

```python
import pandas as pd

# Hàm để mô tả tập dữ liệu
def describe_dataset(file_path):
    # Đọc dữ liệu từ file CSV
    data = pd.read_csv(file_path)
    
    # Số lượng hàng và cột
    num_rows = len(data)
    num_columns = len(data.columns)
    
    # Kiểu dữ liệu của mỗi cột
    data_types = data.dtypes
    
    # Miền giá trị của mỗi cột (nếu là số)
    numeric_ranges = data.describe().loc[['min', 'max']]
    
    # In ra thông tin mô tả
    print("Thông tin mô tả tập dữ liệu:")
    print(f"Số lượng hàng: {num_rows}")
    print(f"Số lượng cột: {num_columns}")
    print("Kiểu dữ liệu của mỗi cột:")
    print(data_types)
    print("Miền giá trị của mỗi cột (nếu là số):")
    print(numeric_ranges)

# Hàm tính toán và hiển thị các thống kê cơ bản
def calculate_basic_stats(file_path):
    # Đọc dữ liệu từ file CSV
    data = pd.read_csv(file_path)
    
    # Tính các thống kê cơ bản
    stats = data.describe()
    
    # In ra các thống kê cơ bản
    print("Các thống kê cơ bản:")
    print(stats)

# Test các hàm
file_path = "data_for_test_10000.csv"
describe_dataset(file_path)
print("\n")
calculate_basic_stats(file_path)
```

```python
import pandas as pd

# Đọc dữ liệu từ file CSV vào DataFrame
data = pd.read_csv("data_for_test_10000.csv")

# Số lượng hàng và cột
num_rows, num_cols = data.shape
print("Số lượng hàng:", num_rows)
print("Số lượng cột:", num_cols)

# Kiểu dữ liệu của mỗi cột
print("\nKiểu dữ liệu của mỗi cột:")
print(data.dtypes)

# Miền giá trị của mỗi cột (nếu là số)
print("\nMiền giá trị của mỗi cột (nếu là số):")
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns
for column in numeric_columns:
    print(f"{column}: {data[column].min()} - {data[column].max()}")


# Hàm tính số lượng hàng
def calculate_count(column):
    return len(column)

# Hàm tính trung bình
def calculate_mean(column):
    return column.mean()

# Hàm tính độ lệch chuẩn
def calculate_std(column):
    return column.std()

# Hàm tính giá trị nhỏ nhất
def calculate_min(column):
    return column.min()

# Hàm tính phân vị 25%
def calculate_25th_quantile(column):
    return column.quantile(0.25)

# Hàm tính phân vị 50%
def calculate_50th_quantile(column):
    return column.quantile(0.5)

# Hàm tính phân vị 75%
def calculate_75th_quantile(column):
    return column.quantile(0.75)

# Hàm tính giá trị lớn nhất
def calculate_max(column):
    return column.max()

# Tính toán các thống kê cơ bản cho cột "Unnamed: 0"
column = data["Unnamed: 0"]
count = calculate_count(column)
mean = calculate_mean(column)
std = calculate_std(column)
min_val = calculate_min(column)
quantile_25 = calculate_25th_quantile(column)
quantile_50 = calculate_50th_quantile(column)
quantile_75 = calculate_75th_quantile(column)
max_val = calculate_max(column)

# Hiển thị các thống kê cơ bản
print("Các thống kê cơ bản:")
print(f"count  {count}")
print(f"mean   {mean}")
print(f"std    {std}")
print(f"min    {min_val}")
print(f"25%    {quantile_25}")
print(f"50%    {quantile_50}")
print(f"75%    {quantile_75}")
print(f"max    {max_val}")
```

```python
# Hàm tính giá trị nhỏ nhất của mỗi cột
def calculate_min_value(column):
    return column.min()

# Hàm tính giá trị lớn nhất của mỗi cột
def calculate_max_value(column):
    return column.max()

# Lặp qua từng cột và tính toán miền giá trị
print("Miền giá trị của mỗi cột:")
for column_name in data.columns:
    if data[column_name].dtype in ['int64', 'float64']:
        min_val = calculate_min_value(data[column_name])
        max_val = calculate_max_value(data[column_name])
        print(f"{column_name}: {min_val} - {max_val}")
    else:
        print(f"{column_name}: Không phải kiểu dữ liệu số")
```
