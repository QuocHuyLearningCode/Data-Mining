```python
import pandas as pd
#Doc du lieu tu file .csv vao mot DataFrame
df = pd.read_csv('C:/Users/LE QUANG HUY/Downloads/data_for_test_10000.csv')
#Hien thi mot so dong dau tien de xem du lieu
print('Du lieu goc: ')
print(df.head())
# Hiển thị thông tin cơ bản về số lượng hàng và cột
print("Số lượng hàng:", len(df))
print("Số lượng cột:", len(df.columns))
print()
# Hiển thị kiểu dữ liệu của mỗi cột
print("Kiểu dữ liệu của mỗi cột:")
print(df.dtypes)
print()
# Hiển thị miền giá trị của mỗi cột
print("Miền giá trị của mỗi cột:")
for column in df.columns:
    print(column + ":", df[column].min(), "-", df[column].max())
def basic_statistics(df):
    numerical_columns = df.select_dtypes(include=['number'])
    stats = {
        'count': numerical_columns.count(),
        'mean': numerical_columns.mean(),
        'std': numerical_columns.std(),
        'min': numerical_columns.min(),
        '25%': numerical_columns.quantile(0.25),
        '50%': numerical_columns.quantile(0.5),
        '75%': numerical_columns.quantile(0.75),
        'max': numerical_columns.max(),
    }
    return pd.DataFrame(stats).transpose()

# Hiển thị kết quả thống kê 
print(basic_statistics(df))
```

# Yêu cầu 3:

```python
import pandas as pd
import matplotlib.pyplot as plt

def load_data(file_path):
    """Load data from a CSV file into a DataFrame."""
    return pd.read_csv('C:/Users/LE QUANG HUY/Downloads/data_for_test_10000.csv')

def display_head(dataframe):
    """Display the first few rows of the DataFrame."""
    print('Du lieu goc: ')
    print(dataframe.head())

def plot_gender_distribution(dataframe):
    """Plot a pie chart to show the gender distribution."""
    plt.figure(figsize=(8, 6))
    dataframe['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['blue', 'pink'])
    plt.title('Phân phối giới tính')
    plt.ylabel('')
    plt.show()

def plot_salary_distribution(dataframe):
    """Plot a histogram to show the salary distribution."""
    plt.figure(figsize=(8, 6))
    plt.hist(dataframe['Salary'], bins=20, color='skyblue', edgecolor='black')
    plt.title('Phân bố mức lương')
    plt.xlabel('Mức lương')
    plt.ylabel('Số lượng')
    plt.show()

def process_birthday_column(dataframe):
    """Convert the 'Birthday' column to datetime format."""
    dataframe['Birthday'] = pd.to_datetime(dataframe['Birthday'], errors='coerce')

def calculate_age(dataframe):
    """Calculate the age until 01/01/2024 and add it as a new column."""
    dataframe['Age'] = (pd.to_datetime('2024-01-01') - dataframe['Birthday']).dt.days // 365

def plot_age_distribution(dataframe):
    """Plot a histogram to show the age distribution."""
    plt.figure(figsize=(8, 6))
    plt.hist(dataframe['Age'], bins=20, color='skyblue', edgecolor='black')
    plt.title('Phân phối độ tuổi tính đến 01/2024')
    plt.xlabel('Tuổi')
    plt.ylabel('Số lượng')
    plt.grid(True)
    plt.show()

def plot_hometown_distribution(dataframe):
    """Plot a bar chart to show the hometown distribution."""
    hometown_counts = dataframe['Hometown'].value_counts()
    plt.figure(figsize=(10, 6))
    hometown_counts.plot(kind='bar', color='skyblue')
    plt.title('Phân phối tỉ lệ hometown')
    plt.xlabel('Hometown')
    plt.ylabel('Số lượng')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Load data
df = load_data('data_cleaned.csv')

# Display the first few rows
display_head(df)

# Plot gender distribution
plot_gender_distribution(df)

# Plot salary distribution
plot_salary_distribution(df)

# Process birthday column
process_birthday_column(df)

# Calculate age
calculate_age(df)

# Plot age distribution
plot_age_distribution(df)

# Plot hometown distribution
plot_hometown_distribution(df)
```
