```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Đọc file CSV và hiển thị dữ liệu
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, header=None, names=columns)

print("Dữ liệu Iris:")
print(df.head())  # Hiển thị 5 dòng đầu

# 2. Kiểu dữ liệu các cột
print("\nKiểu dữ liệu các cột:")
print(df.dtypes)

# 3. Hiển thị các giá trị thống kê
print("\nCác giá trị thống kê:")
print(df.describe())

# 4. Trực quan hóa dữ liệu bằng histogram
for column in df.columns[:-1]:  # Loại cột 'species' vì không phải kiểu số
    plt.figure(figsize=(6, 4))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
Các yêu cầu chính như sau:

1. Đọc file csv và hiển thị dữ liệu

2. Cho biết kiểu dữ liệu các cột

3. Hiển thị các giá trị thống kê (count, mean, std,...)

4. Trực quan hóa lần lượt các thuộc tính dùng các dạng đồ thị histogram

*Yêu cầu: nộp file ipynb và PDF; có thông tin sinh viên trong nội dung
```
