```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Tải dữ liệu (giả sử dữ liệu đã có sẵn trong file CSV)
df = pd.read_csv('D:/download/smsspamcollection/SMSSpamCollection.csv', sep='\t', header=None, encoding='latin-1')

df.columns = ['label', 'message'] 

# Chuyển đổi cột "label" thành 0 (ham) và 1 (spam)
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.3, random_state=42)

# Vector hóa dữ liệu sử dụng CountVectorizer và TfidfVectorizer
count_vectorizer = CountVectorizer(stop_words='english')
X_train_count = count_vectorizer.fit_transform(X_train)
X_test_count = count_vectorizer.transform(X_test)

tfidf_vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Mô hình phân loại Naive Bayes và SVM
models = {
    "Naive Bayes (CountVectorizer)": MultinomialNB(),
    "Naive Bayes (TfidfVectorizer)": MultinomialNB(),
    "SVM (CountVectorizer)": SVC(),
    "SVM (TfidfVectorizer)": SVC(),
}

# Huấn luyện và đánh giá các mô hình
results = {}

for model_name, model in models.items():
    # Dựa vào tên mô hình để quyết định vectorizer sử dụng
    if 'CountVectorizer' in model_name:
        X_train_vec = X_train_count
        X_test_vec = X_test_count
    else:
        X_train_vec = X_train_tfidf
        X_test_vec = X_test_tfidf

    # Huấn luyện mô hình và dự đoán
    model.fit(X_train_vec, y_train)
    y_pred = model.predict(X_test_vec)
    
    # Tính toán độ chính xác và báo cáo phân loại
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    results[model_name] = (accuracy, report)

# In kết quả
for model_name, (accuracy, report) in results.items():
    print(f"\n{model_name}:")
    print(f"Accuracy: {accuracy:.4f}")
    print("Classification Report:")
    print(report)
```
